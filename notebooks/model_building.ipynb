```python
import pandas as pd
```

```python
import statsmodels.api as sm
```

```python
new_dataframe1 = pd.read_csv('..\\data\\new_dataframe1.csv')

new_dataframe1
```

```python
print(new_dataframe1.head())
print(new_dataframe1.describe())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Scatter plot
plt.figure(figsize=(8, 6))
sns.scatterplot(data=new_dataframe1, x='rating', y='distance_x')
plt.title('Scatter Plot of distance vs rating')
plt.xlabel('rating')
plt.ylabel('distance_x')
plt.show()

```

```python
# Box plot
plt.figure(figsize=(10, 6))
sns.boxplot(data=new_dataframe1, x='rating', y='distance_x')
plt.title('Box Plot of distance vs rating')
plt.xlabel('rating')
plt.ylabel('distance_x')
plt.show()
```

```python
# Histogram for 'distance'
plt.figure(figsize=(10, 6))
sns.histplot(data=new_dataframe1, x='distance_x', kde=True, color='green', bins=20)
plt.title('Histogram of Distance')
plt.xlabel('Distance')
plt.ylabel('Frequency')
plt.show()
```

```python
# Histogram for 'rating'
plt.figure(figsize=(10, 6))
sns.histplot(data=new_dataframe1, x='rating', kde=True, color='blue', bins=20)
plt.title('Histogram of Rating')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.show()
```

```python
# Define independent variable (X) and dependent variable (y)
X = new_dataframe1['distance_y']
y = new_dataframe1['rating']

# Adding constant
X = sm.add_constant(X)
```

Provide model output and an interpretation of the results. 

```python
# Creating the regression model
model = sm.OLS(y, X)
results = model.fit()

# Print the summary of the regression model
print(results.summary())
```
